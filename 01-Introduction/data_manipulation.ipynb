```python
def clean_data(df):
    df.columns= df.columns.str.lower()
    strings = list(df.dtypes[df.dtypes == 'object'].index)

    for col in strings:
        df[col] = df[col].str.lower().str.replace(' ', '_')
    return df
```

```python
sch  = clean_data(sch)
bike = clean_data(bike)
```

```python
# function to find the primary key in the data set 
def find_primary_key_candidates(df):
    pk_candidates = []
    
    for col in df.columns:
        if df[col].isna().sum() == 0 and df[col].is_unique:
            pk_candidates.append(col)

    return pk_candidates
```

```python
find_primary_key_candidates(sch)
```

```python
find_primary_key_candidates(bike)
```
